```python
# GPT is called using the call_GPT function. The input in the first step contains a list of the available models. Those are imported as strings and then stored in a list.





# # IMs with 4o mini with code interpreter:
# MAXRANGE = 10
# LLM_model = "gpt-4o-mini-2024-07-18"
# used_prompt = "with_code_interpreter" #"without_CI"
# step_name = "generate_IM"
# prompt = PM.prompt_IM_code_interpreter



# # MSO set with 4o mini with code interpreter:
# MAXRANGE = 10
# LLM_model = "gpt-4o-mini-2024-07-18"
# used_prompt = "with_code_interpreter" #"without_CI"
# step_name = "generate_MSOs"
# prompt = PM.prompt_MSO_code_interpreter



# # IMs with o1 without code interpreter:
# MAXRANGE = 1
# LLM_model = "o1-2024-12-17"
# used_prompt = "without_CI"
# step_name = "generate_IM"
# prompt = PM.prompt_IM_without_CI


# # MSOs with o1 without code interpreter:
# MAXRANGE = 1
# LLM_model = "o1-2024-12-17"
# used_prompt = "without_CI"
# step_name = "generate_MSOs"
# prompt = PM.prompt_MSO_without_CI



# # IMs with o3 mini without code interpreter:
# MAXRANGE = 10
# LLM_model = "o3-mini-2025-01-31"
# used_prompt = "without_CI"
# step_name = "generate_IM"
# prompt = PM.prompt_IM_without_CI


# MSOs with o3 without code interpreter:
MAXRANGE = 10
LLM_model = "o3-mini-2025-01-31"
used_prompt = "without_CI"
step_name = "generate_MSOs"
prompt = PM.prompt_MSO_without_CI






models = dict()
for model_file in os.listdir(model_directory):
    input_file = os.path.join(model_directory, model_file)
    with open(input_file, 'r') as file:
        model_def = file.read()
        models[input_file.split("/")[-1].strip().split("_")[-1].split(".")[0]] = model_def

modelslist = sorted(models.items())
models = dict()
for key, value in modelslist:
    models[key] = value



IMs = dict()
for IM_file in os.listdir(IM_directory):
    input_IM_file = os.path.join(IM_directory, IM_file)
    with open(input_IM_file, 'r') as file:
        incidence_matrix = file.read()
        IMs[os.path.basename(input_IM_file).split("_")[-1].split(".")[0]] = incidence_matrix


IM_list = sorted(IMs.items())
IMs = dict()
for key, values in IM_list:
    IMs[key] = values


if step_name == "generate_IM":
    examples = models
elif step_name == "generate_MSOs":
    examples = IMs
else:
    raise ValueError("Input imported from files is not set properly. Check step_name.")



for repcounter in range(0,MAXRANGE):
    gpt = ThreadsWithAI(prompt, LLM_model)
    client = gpt.client
    assistant_id = gpt.assistantID
    gpt_model_name = gpt.model_name
    thread = gpt.thread





    #Prepare output file
    GPT_model_name_sanitized = gpt_model_name.replace("-", "_")
    results_dir = "results"
    if not os.path.exists(results_dir):
        os.makedirs(results_dir)
    dt = datetime.now().strftime("%Y%m%d_%H%M%S")
    output_file = os.path.join(
            results_dir, 
            f"{GPT_model_name_sanitized}_{step_name}_{used_prompt}_rep{repcounter}_{dt}.txt"
        )
    f = open(output_file, "w")
    for name, ex in examples.items():

        response = gpt.call_GPT(thread,ex, client, assistant_id)

        # Print responses
        print("\nGPT Analysis  "+str(name)+":")
        print(response)

        
        f.write("Name: "+ name +"\n")
        f.write(str(response))
        f.write("\n\n")
        f.flush()

    f.close()





```
